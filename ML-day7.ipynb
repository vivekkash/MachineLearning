```python
def palindrome(num):
    num = str(num)
    if num == num[::-1]:
        return num+" True"
    else:
        num = int(num)+1
        return palindrome(num)
    
```

```python
palindrome(121)
```

```python
palindrome(123)
```

```python
def a_scramble(str1,str2):
    for x in str2:
        if x.lower() not in str1.lower():
            return False
    return True    
```

```python
a_scramble("Tom Marvolo Riddle","Voldemort")
```

```python
a_scramble("ticket","chat")
```

```python
from math import sqrt

def isPerfectSquare(x):
    s = int(sqrt(x))
    return s*s == x

def check_fib(num):
    return isPerfectSquare(5 * num * num + 4) or isPerfectSquare(5 * num * num - 4)
       
        
```

```python
check_fib(377)
```

```python
def compress(word):
    temp = {}
    compression = ''
    for x in word:
        if x in temp.keys():
            temp[x] = temp[x] + 1
        else:
            temp[x] = 1   
    for (key, value) in temp.items():
        compression = compression + (key+str(value))
    return compression
```

```python
compress('xxcccdex')
```

```python
def compress_successive(word):
        length_word = len(word)
        count = 1
        s = ""
        for i in range(length_word-1):
            if word[i] == word[i+1]:
                count = count + 1
                if i + 1 == length_word-1:
                    s = s + word[i+1] + str(count)
            else:
                s = s + word[i]+str(count)
                count = 1
                if i + 1 == length_word-1:
                    s = s + word[i+1] + str(count)

            #print(s)
        return s
```

```python
compress_successive('xxxxxnnnnxxxmmmmwwwwax')
```

```python
#Code starts here

#Function to compress string
def compress_3(word):
    word=word.lower()
    mist=[]
    l=0
    while(l<len(word)):
        m=word[l]
        j=0
        while(l<len(word) and word[l]==m):
                j=j+1
                l=l+1

        mist.append(m)
        mist.append(str(j))

    return ''.join(mist)
```

```python
compress_3('xxxxxnnnnxxxmmmmwwwwax')
```
