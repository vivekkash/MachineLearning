```python
#nCr*pr*qn-r/(n-r)!*r!

import math
def ncr(n,r,p):
    q = 1-p
    numerator = math.factorial(n)*(p**r)*(q**(n-r))
    denominator = (math.factorial(n-r))*math.factorial(r)
    return numerator/denominator
```

```python
ncr(11,7,0.39)
```

```python
#prob of getting bike rented 8 days out of 10

#n=10
#r=8
ncr(10,8,prob_4000_bike_rented)


#prob of getting bike rented 10 days out of 10

ncr(10,10,prob_4000_bike_rented)

```

```python
# KNN using scilearn
df = pd.read_csv('/Users/03000070/Downloads/diabetes.csv')
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
df.shape
```

```python
X = df.drop(["diabetes"],1)
y = df["diabetes"] 
```
